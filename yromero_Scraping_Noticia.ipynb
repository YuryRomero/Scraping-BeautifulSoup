```python
#ETAPA DE ADQUISICION DE DATOS
#Tecnina de Scraping

from urllib.request import FancyURLopener
from bs4 import BeautifulSoup
import pandas as pd

```

```python
import sys
from PyQt5.QtWidgets import * 
from PyQt5.QtCore import *
from PyQt5.QtWebKit import *
from PyQt5.QtWebKitWidgets import *
```

```python
class Client(QWebPage):
    def __init__(self,url):
        self.app = QApplication(sys.argv)
        QWebPage.__init__(self)
        self.loadFinished.connect(self.on_page_load)
        self.mainFrame().load(QUrl(url))
        self.app.exec_()
        
    def on_page_load(self):
        self.app.quit()
```

```python
class MyOpener(FancyURLopener):
    version = 'Mozilla/5.0 (Windows; U; Windows NT 5.1; it; rv:1.8.1.11) Gecko/20071127 Firefox/2.0.0.11'
myopener = MyOpener()
```

```python
def is_the_only_string_within_a_tag(s):
    """Return True if this string is the only child of its parent tag."""
    return (s == s.parent.string)
```

```python
#scraping peru 21
web = 'url_pagina_noticia'#colocar la url de la pagina
URL=web+'/archivo/todas/2017-12-10'
#numero de dias noticias
nume_dias = 2
exis_pagina = True
i=0
file = open("data_noticia_actual.txt","w",encoding='utf-8')        
while exis_pagina :
    print(URL)
    #abrimos la pagina de noticias
    html=myopener.open(URL).read()
    soup=BeautifulSoup(html,"lxml")
    #obtenemos lista de enlaces de noticias
    list_noticia = soup.find_all(lambda tag: tag.name == 'a' and tag.get('class') == ['page-link'] and not tag.has_attr('id'))
    #recorrer lista de enlaces
    for link_noticia in list_noticia:
        #abrimos pagina de noticia
        url_noticia  = web+link_noticia.get("href")
        html_noticia = myopener.open(url_noticia).read()
        soup_noticia = BeautifulSoup(html_noticia,"lxml")
        
        print(url_noticia)
        #lista_texto = soup_noticia.find_all(string=is_the_only_string_within_a_tag)
        #obtenemos los parrafos de la noticia
        lista_tag = soup_noticia.find_all("p",class_="parrafo first-parrafo")
        
        #buscamos texto de noticia y guardamos en archivos texto
        for tag in lista_tag :
            file.write(tag.getText()+'\n')
        #break
        file.write('\n\n')
       
    i += 1
    
    #Verificar siguiente pagina
    sigu_pagina=(soup.find_all('li',class_="date-item date-prev",limit=1))[0].find_next()
    
    #print(sigu_pagina[0].find_next())
    print(web+sigu_pagina.get("href"))
    
    
    if len(sigu_pagina) != 0 :
        URL  = web+sigu_pagina.get("href")
    else :
        exis_pagina = False
    
    if i==nume_dias:
        break

file.close()
    
```
